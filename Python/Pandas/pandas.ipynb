# Merge, Join e Concatenao

```python
import pandas as pd

df1 = pd.read_csv(r"G:\My Drive\Estudos_Cursos\UltimateDA\Python\Pandas\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"G:\My Drive\Estudos_Cursos\UltimateDA\Python\Pandas\LOTR 2.csv")
df2
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
df1.merge(df2, how='cross')
```

```python
df1.join(df2, on='FellowshipID', how='outer', lsuffix='_Left', rsuffix='_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_Left', rsuffix='_Right', how='outer')
df4
```

```python
pd.concat([df1, df2], join='outer', axis=1)
```

# Limpeza de Dados

```python
import pandas as pd

df = pd.read_excel(r"G:\My Drive\Estudos_Cursos\UltimateDA\Python\Pandas\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
# df = df.drop(columns='Not_Useful_Column')
df
```

```python
# df['Last_Name'] = df['Last_Name'].str.lstrip('...')
# df['Last_Name'] = df['Last_Name'].str.lstrip('/')
# df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df['Last_Name'] = df['Last_Name'].str.rstrip('123._/')
df
```

```python
# df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
# df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] + '-')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na', '', regex=True)
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df = df.replace('N/a', '')
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
df
```

```python
df = df.reset_index(drop=True)
df
```
